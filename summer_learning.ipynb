```python
import gradio as gr
import yaml
from typing import List, Tuple, Iterator
from dataclasses import dataclass, field
from enum import Enum
from PIL import Image

class Mode(Enum):
    HORIZONTAL = "水平"
    VERTICAL = "垂直"

class ParticleType(Enum):
    ANGRY_VILLAGER = "ANGRY_VILLAGER (怒った村人)"
    ASH = "ASH (灰)"
    BARRIER = "BARRIER (バリア)"
    BLOCK = "BLOCK (ブロック)"
    BLOCK_CRACK = "BLOCK_CRACK (ブロック破壊)"
    BUBBLE = "BUBBLE (泡)"
    BUBBLE_COLUMN_UP = "BUBBLE_COLUMN_UP (上昇する泡柱)"
    CAMPFIRE_COSY_SMOKE = "CAMPFIRE_COSY_SMOKE (キャンプファイヤーの煙(小))"
    CAMPFIRE_SIGNAL_SMOKE = "CAMPFIRE_SIGNAL_SMOKE (キャンプファイヤーの煙(大))"
    CHERRY_LEAVES = "CHERRY_LEAVES (サクラの葉)"
    CLOUD = "CLOUD (雲)"
    CRIMSON_SPORE = "CRIMSON_SPORE (真紅の胞子)"
    CRIT = "CRIT (クリティカル)"
    DOLPHIN = "DOLPHIN (イルカ)"
    DRAGON_BREATH = "DRAGON_BREATH (ドラゴンブレス)"
    DRIPPING_HONEY = "DRIPPING_HONEY (滴る蜂蜜)"
    DRIPPING_LAVA = "DRIPPING_LAVA (滴る溶岩)"
    DRIPPING_WATER = "DRIPPING_WATER (滴る水)"
    DUST = "DUST (塵)"
    DUST_COLOR_TRANSITION = "DUST_COLOR_TRANSITION (色が変化する塵)"
    DUST_PLUME = "DUST_PLUME (ほこり)"
    ELECTRIC_SPARK = "ELECTRIC_SPARK (電気の火花)"
    ENCHANTED_HIT = "ENCHANTED_HIT (エンチャントヒット)"
    END_ROD = "END_ROD (エンドロッド)"
    EXPLOSION = "EXPLOSION (爆発)"
    EXPLOSION_EMITTER = "EXPLOSION_EMITTER (爆発発生源)"
    FALLING_DUST = "FALLING_DUST (落下する塵)"
    FALLING_HONEY = "FALLING_HONEY (落下する蜂蜜)"
    FALLING_LAVA = "FALLING_LAVA (落下する溶岩)"
    FALLING_NECTAR = "FALLING_NECTAR (落下する蜜)"
    FALLING_WATER = "FALLING_WATER (落下する水)"
    FIREWORK = "FIREWORK (花火)"
    FLAME = "FLAME (炎)"
    FLASH = "FLASH (閃光)"
    GLOW = "GLOW (発光)"
    GLOW_SQUID_INK = "GLOW_SQUID_INK (発光イカスミ)"
    HEART = "HEART (ハート)"
    ITEM_SLIME = "ITEM_SLIME (スライムアイテム)"
    ITEM_SNOWBALL = "ITEM_SNOWBALL (雪玉アイテム)"
    LANDING_HONEY = "LANDING_HONEY (着地する蜂蜜)"
    LANDING_LAVA = "LANDING_LAVA (着地する溶岩)"
    LAVA = "LAVA (溶岩)"
    MOB_APPEARANCE = "MOB_APPEARANCE (モブ出現)"
    NAUTILUS = "NAUTILUS (オウムガイ)"
    NOTE = "NOTE (音符)"
    POOF = "POOF (煙)"
    PORTAL = "PORTAL (ポータル)"
    RAIN = "RAIN (雨)"
    REVERSE_PORTAL = "REVERSE_PORTAL (逆ポータル)"
    SCRAPE = "SCRAPE (削り)"
    SCULK_CHARGE = "SCULK_CHARGE (スカルクチャージ)"
    SCULK_CHARGE_POP = "SCULK_CHARGE_POP (スカルクチャージポップ)"
    SCULK_SOUL = "SCULK_SOUL (スカルクソウル)"
    SHRIEK = "SHRIEK (悲鳴)"
    SMALL_FLAME = "SMALL_FLAME (小さな炎)"
    SMOKE = "SMOKE (煙)"
    SNEEZE = "SNEEZE (くしゃみ)"
    SNOWFLAKE = "SNOWFLAKE (雪片)"
    SOUL = "SOUL (ソウル)"
    SOUL_FIRE_FLAME = "SOUL_FIRE_FLAME (ソウルファイアの炎)"
    SPIT = "SPIT (唾)"
    SPORE_BLOSSOM_AIR = "SPORE_BLOSSOM_AIR (胞子の花の空気)"
    SQUID_INK = "SQUID_INK (イカスミ)"
    SWEEP_ATTACK = "SWEEP_ATTACK (薙ぎ払い攻撃)"
    TOTEM_OF_UNDYING = "TOTEM_OF_UNDYING (不死のトーテム)"
    UNDERWATER = "UNDERWATER (水中)"
    VIBRATION = "VIBRATION (振動)"
    WARPED_SPORE = "WARPED_SPORE (歪んだ胞子)"
    WAX_OFF = "WAX_OFF (ワックスオフ)"
    WAX_ON = "WAX_ON (ワックスオン)"
    WHITE_ASH = "WHITE_ASH (白い灰)"
    WITCH = "WITCH (魔女)"

@dataclass
class ParticleData:
    type: ParticleType
    color: str = "#FFFFFF"
    size: float = 1.0
    material: str = "STONE"

@dataclass
class ImageDetails:
    name: str
    width: int
    height: int
    pixel_colors: List[List[Tuple[int, ...]]]

    @classmethod
    def from_image(cls, image: Image.Image, name: str) -> "ImageDetails":
        if image.mode != 'RGBA':
            image = image.convert('RGBA')
        color_access = image.load()
        return cls(
            name=name,
            width=image.width,
            height=image.height,
            pixel_colors=[
                [color_access[x, y] for x in range(image.width)] for y in range(image.height)
            ]
        )

@dataclass
class CodeGenerator:
    mode: Mode
    particle_data: ParticleData
    particle_interval: float
    base_forward_offset: float
    base_side_offset: float
    base_y_offset: float
    image: ImageDetails

    forward_offset: float = field(init=False)
    side_offset: float = field(init=False)
    y_offset: float = field(init=False)
    def __post_init__(self):
      if self.mode is Mode.HORIZONTAL:
        # forward_offsetをパーティクル間隔に基づいて調整
        self.forward_offset = self.base_forward_offset + self.image.width * self.particle_interval / 2
        # side_offsetを側面に基づいて調整
        self.side_offset = self.base_side_offset
        # y_offsetを基づいて調整（Y軸オフセット）
        self.y_offset = self.image.height * self.particle_interval + self.base_y_offset
      elif self.mode is Mode.VERTICAL:
        # 縦向きの場合
        self.forward_offset = self.base_forward_offset
        self.side_offset = self.image.width * self.particle_interval / 2 + self.base_side_offset
        self.y_offset = self.base_y_offset


    def generate_code(self) -> Iterator[str]:
        yield f"{self.image.name}: \n  Skills: \n"
        for y_index, colors_at_y in enumerate(self.image.pixel_colors):
            for x_index, color_at_xy in enumerate(colors_at_y):
                if len(color_at_xy) < 4 or color_at_xy[3] == 0:
                    continue
                yield self._generate_particle_effect(x_index, y_index, color_at_xy)

    def _generate_particle_effect(self, x_index: int, y_index: int, color_at_xy: Tuple[int, ...]) -> str:
        particle_type = self.particle_data.type.name
        color = "#{0:02x}{1:02x}{2:02x}".format(color_at_xy[0], color_at_xy[1], color_at_xy[2])
        size = round(color_at_xy[3] / 255 * self.particle_data.size, 3) if len(color_at_xy) > 3 else self.particle_data.size

        effect = f"    - effect:particles{{particle={particle_type};"

        if particle_type in ["BLOCK", "BLOCK_CRACK", "FALLING_DUST"]:
            effect += f"material={self.particle_data.material};"
        elif particle_type not in ["BARRIER", "LIGHT", "VIBRATION"]:
            effect += f"color={color};"

        if particle_type in ["DUST", "DUST_COLOR_TRANSITION"]:
            effect += f"size={size};"

        effect += "amount=1;"

        fo = round(self.forward_offset + y_index * self.particle_interval, 3) if self.mode is Mode.VERTICAL else round(self.forward_offset, 3)
        so = round(self.side_offset - x_index * self.particle_interval, 3)
        y = round(self.y_offset, 3) if self.mode is Mode.VERTICAL else round(self.y_offset - y_index * self.particle_interval, 3)

        effect += f"forwardOffset={fo};sideOffset={so};yOffset={y}}}\n"

        return effect

def generate_particle_code(
    image: Image.Image,
    mode: str,
    particle_type: str = "FLAME (炎)",
    particle_interval: float = 0.1,
    particle_size: float = 1.0,
    base_forward_offset: float = 0,
    base_side_offset: float = 0,
    base_y_offset: float = 2,
    particle_color: str = "#FFFFFF",
    particle_material: str = "STONE"
) -> Tuple[str, str]:
    if image is None:
        return yaml.dump({
            "error": "画像がアップロードされていません。",
            "image_info": "画像情報: なし",
            "particle_count": 0,
        }, allow_unicode=True), None

    image_details = ImageDetails.from_image(image, "アップロードされた画像")

    particle_enum = next((p for p in ParticleType if p.value.startswith(particle_type.split(" ")[0])), None)
    if particle_enum is None:
        return yaml.dump({
            "error": f"無効なパーティクルタイプ: {particle_type}",
            "image_info": "画像情報: なし",
            "particle_count": 0,
        }, None, allow_unicode=True)

    particle_data = ParticleData(
        type=particle_enum,
        color=particle_color,
        size=particle_size,
        material=particle_material
    )

    generator = CodeGenerator(
        mode=Mode[mode],
        particle_data=particle_data,
        particle_interval=particle_interval,
        base_forward_offset=base_forward_offset,
        base_side_offset=base_side_offset,
        base_y_offset=base_y_offset,
        image=image_details
    )

    code_lines = [line for line in generator.generate_code() if line.strip().startswith("- effect:particles")]

    if not code_lines:
        return yaml.dump({
            "error": "適切なパーティクル効果が生成されませんでした。",
            "image_info": f"画像サイズ: {image_details.width}x{image_details.height}",
            "particle_count": 0,
        }, allow_unicode=True), None

    yaml_code = "\n".join(code_lines)
    particle_count = len(code_lines)

    result = {
        "code": yaml_code,
        "image_info": f"画像サイズ: {image_details.width}x{image_details.height}",
        "particle_count": particle_count,
    }

    # Save to YML file for download
    yml_file_path = "/mnt/data/particle_code.txt"
    with open(yml_file_path, "w", encoding="utf-8") as file:
        file.write(yaml_code)

    return yaml_code, yml_file_path

# Adjust Gradio interface to make sure it uses this function



def update_particle_type_choices(mode):
    return gr.Dropdown(choices=[p.value for p in ParticleType], value="FLAME (炎)")

with gr.Blocks() as demo:
    gr.Markdown("# Minecraft パーティクルコードジェネレーター")
    gr.Markdown("画像をアップロードし、パラメータを設定してMinecraftのパーティクルコードを生成します。")

    with gr.Row():
        with gr.Column():
            image_input = gr.Image(type="pil", label="画像をアップロード")
            mode = gr.Radio(["HORIZONTAL", "VERTICAL"], label="モード", value="HORIZONTAL")
            particle_type = gr.Dropdown(choices=[p.value for p in ParticleType], label="パーティクルタイプ", value="FLAME (炎)")
            particle_interval = gr.Slider(minimum=0.01, maximum=1.0, step=0.01, value=0.1, label="パーティクル間隔")
            particle_size = gr.Slider(minimum=0.1, maximum=5.0, step=0.1, value=1.0, label="パーティクルサイズ")
            particle_color = gr.ColorPicker(label="パーティクル色", value="#FFFFFF")
            particle_material = gr.Dropdown(choices=["STONE", "DIRT", "GRASS_BLOCK"], label="ブロック材質", value="STONE")
            base_forward_offset = gr.Number(value=0, label="前方オフセット")
            base_side_offset = gr.Number(value=0, label="側面オフセット")
            base_y_offset = gr.Number(value=2, label="Y軸オフセット")
            generate_button = gr.Button("コード生成")

        with gr.Column():
            output_text = gr.TextArea(label="生成されたコード", interactive=False)
            info_text = gr.TextArea(label="情報", interactive=False)
            download_button = gr.File(label="コードをダウンロード")

    generate_button.click(
        generate_particle_code,
        inputs=[
            image_input,
            mode,
            particle_type,
            particle_interval,
            particle_size,
            base_forward_offset,
            base_side_offset,
            base_y_offset,
            particle_color,
            particle_material
        ],
        outputs=[output_text, download_button]
    )

    mode.change(
        update_particle_type_choices,
        inputs=[mode],
        outputs=[particle_type]
    )

demo.launch()
```
